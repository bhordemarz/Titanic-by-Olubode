```python
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
titanic_train=pd.read_csv('train.csv') #import the train data
titanic_train
```

```python
titanic_test=pd.read_csv('test.csv') #import the test data
titanic_test
```

```python
titanic_train.info() #display information of the train data
```

```python
#convert the column "Sex" from strings to integer
gender = {'male': 1,'female': 0}   
titanic_test['Sex'] = [gender[item] for item in titanic_test['Sex']]
titanic_train['Sex'] = [gender[item] for item in titanic_train['Sex']]
```

```python
titanic_train.info() #display information of the train data
```

```python
titanic_test.info() #display information of the test data
```

```python
#select columns for the input training data
X_train=titanic_train.drop(columns=['PassengerId', 'Fare', 'Embarked', 'Survived', 'Age', 'Name', 'Ticket', 'Cabin'])
X_train
```

```python
#select columns for the output training data
Y_train=titanic_train['Survived']
Y_train
```

```python
#select columns for the input test data
X_test=titanic_test.drop(columns=['PassengerId','Fare', 'Embarked', 'Age', 'Name', 'Ticket', 'Cabin'])
X_test
```

```python
model=DecisionTreeClassifier() #model your data
```

```python
model.fit(X_train, Y_train) #train yyour model
predictions = model.predict(X_test) #predict using your test data
```

```python
#convert your output test data to csv file
output = pd.DataFrame({'PassengerId': titanic_test.PassengerId, 'Survived': predictions})
output.to_csv('Olu_submission.csv', index=False)
print("Your submission was successfully saved!")
```
